```python
from itertools import combinations
import numpy as np
```

## Part 1

```python
def within_borders(pos, size):
    return pos[0]>=0 and pos[0]<=size[0] and pos[1]>=0 and pos[1]<=size[1]
```

```python
def valid_combinations_pt1(coords, size):
    antinodes =[]
    for antenna_pairs in list(combinations(coords,2)):
        a = np.array(antenna_pairs[0])
        b = np.array(antenna_pairs[1])
        if within_borders(2*a -b, size):
            antinodes.append(2*a-b)
        if within_borders(2*b -a, size):
            antinodes.append(2*b-a)
    return antinodes
```

```python
def compute_antinodes_pt1(antennas, size):
    antinodes =[]
    
    for coords in antennas.values():
        antinodes.extend(valid_combinations_pt1(coords, size))
    
    return set([tuple(pos) for pos in antinodes])
```

```python
map = []
antennas={}
with open('Day08_input.txt') as f:
     for row,line in enumerate(f):
        line = line.replace("\n","")
        map.append(line)
        for col, char in enumerate(line):
            if char != '.': 
                if char not in antennas.keys():
                    antennas[char]= [(row, col)]
                else:
                    antennas[char].append((row, col))

size =(row, col)
unique_antinodes= compute_antinodes_pt1(antennas, size)

print(len(unique_antinodes))
```

## Part 2

```python
def find_resonations(a, b, size):
    resons = []
    steps = 0
    a = np.array(a)
    b = np.array(b)
    reson = a 
    while within_borders(reson, size):
        resons.append(reson)
        steps +=1
        reson= a+ steps*(a-b)
    
    return resons
```

```python
def valid_combinations_pt2(coords, size):
    antinodes =[]
    for antenna_pairs in list(combinations(coords,2)):
        a = antenna_pairs[0]
        b = antenna_pairs[1]
        antinodes.extend(find_resonations(a, b, size))
        antinodes.extend(find_resonations(b, a, size))
    return antinodes
```

```python
def compute_antinodes_pt2(antennas, size):
    antinodes =[]
    
    for coords in antennas.values():
        antinodes.extend(valid_combinations_pt2(coords, size))
        
    return set([tuple(pos) for pos in antinodes])
```
