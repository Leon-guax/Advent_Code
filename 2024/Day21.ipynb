```python
# -1 represents the forbidden hole, and 10 the A button
num_keypad = np.array([[7, 8, 9], [4, 5, 6], [1, 2, 3], [-1, 0, 10]])


def num_keypad_instructions(cur_digit, next_digit):
    cur_pos = np.where(num_keypad==cur_digit)
    cur_y, cur_x = cur_pos[0][0], cur_pos[1][0]
    next_pos = np.where(num_keypad==next_digit)
    next_y, next_x = next_pos[0][0], next_pos[1][0]
    
    move = []
    
    # if we start on the leftmost and we go bottom, we do the full horizontal move first
    if cur_x ==0 and next_y ==3:
        diff_x = next_x - cur_x
        move.extend(['>']*diff_x)
        cur_x = next_x
    
    
    # if we start on the bottom and we go leftmost, we do the full vertical move first
    elif cur_y ==3 and next_x ==0:
        diff_y = next_y - cur_y
        move.extend(['^']*abs(diff_y))
        cur_y = next_y

    
    # if diff_x ==0 I extend with an empty list
    diff_x = next_x - cur_x
    if diff_x >0:
        move.extend(['>']*diff_x)
    else:
        move.extend(['<']*abs(diff_x))
    # print(cur_y, cur_x, next_y, next_x)
        
    # if diff_y ==0 I extend with an empty list
    diff_y = next_y - cur_y
    if diff_y >0:
        move.extend(['v']*diff_y)
    else:
        move.extend(['^']*abs(diff_y))

    return move


num_keypad_instructions(10,4)
```

```python
dirs =  {
    "^": np.array((-1, 0)),
    "v": np.array((1, 0)),
    "<": np.array((0,-1)),
    ">": np.array((0, 1))
    }

def test_numpad_string(numpad_str):
    pos = np.array([3, 2])
    output = []
    for idx,move in enumerate(numpad_str):
        if move =="A":
            output.append(str(num_keypad[tuple(pos)]))
        else:
            pos+= dirs[move]  
            
        if np.all(pos ==np.array([3, 0])):
            raise ValueError(f"Passed over the forbidden empty space in numpad at step n {idx+1} = {move}")
        
        output =["A" if n =="10" else n for n in output]
    return "".join(output)    
test_numpad_string("^<<^A>A>AvvA")
    
```

```python
dir_keypad = np.array([["x", "^", "A"], ["<", "v", ">"]])

def dig_keypad_instructions(cur_but, next_but):
    cur_pos = np.where(dir_keypad==cur_but)
    cur_y, cur_x = cur_pos[0][0], cur_pos[1][0]
    next_pos = np.where(dir_keypad==next_but)
    next_y, next_x = next_pos[0][0], next_pos[1][0]

    move = []
    # if we are leftmost and we go top
    if cur_x ==0 and next_y ==0:
        diff_x = next_x - cur_x
        move.extend(['>']*diff_x)
        cur_x = next_x
        
    # if we are top and we go leftmost, we do 1 step max
    if cur_y ==0 and next_x ==0:
        move.append('v')
        cur_y =1
        
    # if diff_y ==0 I extend with an empty list
    diff_y = next_y - cur_y
    if diff_y >0:
        move.extend(['v']*diff_y)
    else:
        move.extend(['^']*abs(diff_y))
    
    # if diff_x ==0 I extend with an empty list
    diff_x = next_x - cur_x
    if diff_x >0:
        move.extend(['>']*diff_x)
    else:
        move.extend(['<']*abs(diff_x))

    # print(move)
    # print(diff_y, diff_x)
    return move


dig_keypad_instructions("A", "<")
```

```python
def test_digpad_string(digpad_str):
    pos = np.array([0, 2])
    output = []

    for idx, move in enumerate(digpad_str):
        # print('- ', idx, move, pos)
        if move =="A":
            output.append(str(dir_keypad[tuple(pos)]))
        else:
            pos+= dirs[move]  
        if np.all(pos ==np.array([0, 0])):
            raise ValueError(f"Passed over the forbidden empty space in numpad at step n {idx+1} = {move}")

    return "".join(output)
print("^<<^A>A>AvvA")
print(test_digpad_string("<Av<AA>^A>AvA^AvA^A<vAA>^A"))
    
```

```python
list_instr = []


with open('Day21_input.txt') as f:
     for line in f:
        list_instr.append(line.strip())


# list_instr = [
#             "029A",
#             "980A",
#             "179A",
#             "456A",
#             "379A",
#             ]

final_instr = {}
compl_sum = 0
for instr in list_instr:
    num_moves =[]
    cur_dig = 10
    for num in instr:
        if num.isdigit():
            new_dig = int(num)
        else:
            new_dig = 10
        num_moves.extend(num_keypad_instructions(cur_dig, new_dig)+["A"])
        cur_dig = new_dig
    # print("".join(num_moves))
    
    kp1_moves = []
    cur_kp1 = "A"
    for m in num_moves:
        kp1_moves.extend(dig_keypad_instructions(cur_kp1, m) +["A"])
        cur_kp1 = m
    # print("".join(kp1_moves))
    
    kp2_moves = []
    cur_kp2 = "A"
    for m in kp1_moves:
        kp2_moves.extend(dig_keypad_instructions(cur_kp2, m) +["A"])
        cur_kp2 = m
    # print("".join(kp2_moves))
    
    final_instr[instr] =["".join(kp2_moves),len(kp2_moves),int(instr[:-1]), len(kp2_moves)*int(instr[:-1])]
    compl_sum +=len(kp2_moves)*int(instr[:-1])

print()
for instr in list_instr:
    print(instr, final_instr[instr])
    
print()    
print(final_instr[instr][0])
print(compl_sum)
# 94758 too high
```
