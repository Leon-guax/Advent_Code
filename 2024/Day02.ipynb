# Part 1

```python
def is_safe(nrs):
    if not (nrs[-1]>nrs[0]):
        nrs.reverse()     
    diffs = [nrs[idx+1] - nrs[idx] for idx in range(len(nrs)-1)]   
    return max(diffs)<=3 and min(diffs)>0

test = [
[7, 6, 4, 2, 1],
[1, 2, 7, 8, 9],
[9, 7, 6, 2, 1],
[1, 3, 2, 4, 5],
[8, 6, 4, 4, 1],
[1, 3, 6, 7, 9],
[16, 18, 20, 23, 26, 27]]

for report in test:
    print(f"{report} : {is_safe(report)} ")
```

```python
count_safe = 0
with open('Day02_input.txt') as f:
     for line in f:
            report =[int(x) for x in line.split()]
            if is_safe(report):
                count_safe +=1
                

print(count_safe)
```

# Part 2

```python
from collections import Counter

def is_almost_safe(nrs):
    if not (nrs[-1]>nrs[0]):
        nrs.reverse()   
    for idx in range(len(nrs)-1):
        if nrs[idx+1] - nrs[idx] >3 or nrs[idx+1] - nrs[idx]<1:
            pop_idx = nrs[0:idx]+nrs[idx+1:]
            pop_idx_n = nrs[0:idx+1]+nrs[idx+2:]
            print(f"{nrs} --> either {pop_idx} or {pop_idx_n} ")
            
            return is_safe(pop_idx) or is_safe(pop_idx_n)
                        
    return True

test = [
[7, 6, 4, 2, 1],
[1, 2, 7, 8, 9],
[9, 7, 6, 2, 1],
[1, 3, 2, 4, 5],
[8, 6, 4, 4, 1],
[1, 3, 6, 7, 9],
[16, 18, 20, 23, 26, 27],
[15, 15, 16, 15, 19]]
for report in test:
    print(f"{report} --> is_safe: {is_safe(report)},  is_almost_safe: {is_almost_safe(report)}\n\n")
```

```python
count_almost_safe = 0
with open('Day02_input.txt') as f:
     for line in f:
            report =[int(x) for x in line.split()]
            print(f"{report} --> is_safe: {is_safe(report)},  is_almost_safe: {is_almost_safe(report)}\n")
            if is_almost_safe(report):
                count_almost_safe +=1
                

print(count_almost_safe)
```

```python
count_almost_safe = 0
with open('Day02_input.txt') as f:
     for idx, line in enumerate(f):
            report =[int(x) for x in line.split()]
            iss = is_safe(report)
            isas = is_almost_safe(report)
            if iss != isas:
                print(f"---{idx}")
                print(f"{report} --> is_safe: {iss},  is_almost_safe: {isas}\n\n")
```
