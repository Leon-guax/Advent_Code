```python
with open('Day02_input.txt') as f:
    for line in f:
        input = line.split(',')

ranges = []

for part in input:
    a, b = part.split("-")
    start = int(a)
    end = int(b)
    ranges.append(range(start, end + 1))  # inclusive
```

# Part 1

```python
doubles = 0
for rg in ranges:
    for num in rg:
        string = str(num)
        if len(string)%2:
            continue
        else:
            half_l= int(len(string)/2)
            if string[:half_l] == string[half_l:]:
                doubles +=num
                
print(doubles)
```

# Part 2

```python
set_repetitions = set()

verbose = False

for rg in ranges:
    verbose and print('\n\n', rg)
    for num in rg:
        string = str(num)
        l_str = len(string)
        divisors = [d for d in range(1, l_str) if l_str % d == 0]
        verbose and  print('divisors= ', divisors)
        if len(divisors)==0:
            continue
        else:
            for sub_len in divisors:
                verbose and print(f'{string[:sub_len]} * {int(l_str/sub_len)} = {string[:sub_len]*int(l_str/sub_len)} =? {string}')
                if string == string[:sub_len]*int(l_str/sub_len):
                    verbose and print('yes')
                    set_repetitions.add(num)
                
print(sum(set_repetitions))
```
