## Part 2 (clean)

```python
def is_horizontal(edge):
    return edge[0][0]==edge[1][0]
```

```python
def is_horizontal(edge):
    return edge[0][1]==edge[1][1]

def is_crossing(A, BCDE):
    is_A_horizontal = is_horizontal(A)
    
    for ed in BCDE:
        if is_horizontal(ed) != is_A_horizontal:
            if is_A_horizontal:
                Ay = A[0][1]
                Ax1, Ax2 = sorted([A[0][0], A[1][0]])
                edx = ed[0][0]
                edy1, edy2 =sorted([ed[0][1], ed[1][1]])
                if (edy1 < Ay < edy2) and (Ax1 < edx < Ax2):
                    return True 
            else:
                Ax = A[0][0]
                Ay1, Ay2 = sorted([A[0][1], A[1][1]])
                edy = ed[0][1]
                edx1, edx2 =sorted([ed[0][0], ed[1][0]])
                if (edx1 < Ax < edx2) and (Ay1 < edy < Ay2):
                    return True   
    return False



print(is_crossing([[0,0], [0,5]], [ [[1, 1], [1, 2]]] ))
print(is_crossing([[0,0], [0,5]], [ [[0, 0], [0, 3]]] ))
print(is_crossing([[0,0], [0,5]], [ [[-1, 1], [1, 1]]] ))
```

```python
import matplotlib.pyplot as plt

coords =[]
edges =[]
max_area = 0
with open('Day09_input.txt') as f:
     for line in f :
        coords.append([int(coord) for coord in line.rstrip('\n').split(',')])

#also includes the last-first connections
for idx in range(len(coords)):
    edges.append((coords[idx-1], coords[idx]))
    

max_area = 0
for i, j in tqdm(combinations(range(len(coords)), 2)):    
    ci = coords[i]
    cj = coords[j]
    side1 = (ci, [ci[0], cj[1]])     # ci → (x1, y2)
    side2 = ([ci[0], cj[1]], cj)     # (x1, y2) → cj
    side3 = (cj, [cj[0], ci[1]])     # cj → (x2, y1)
    side4 = ([cj[0], ci[1]], ci)     # (x2, y1) → ci 
    square_sides =[side1, side2, side3, side4]

    x = abs(ci[0] -cj[0])+1
    y = abs(ci[1] -cj[1])+1
    area = x*y
    
    crossing = False
    if area > max_area:
        
        xmin = min(ci[0], cj[0])
        xmax = max(ci[0], cj[0])
        ymin = min(ci[1], cj[1])
        ymax = max(ci[1], cj[1])
        
        inside_points = [
        coord for coord in coords 
        if coord != ci and coord != cj
        and xmin < coord[0] < xmax
        and ymin < coord[1] < ymax
        ]
       
        
        for edge in edges:
            if is_crossing(edge, square_sides):
                crossing = True 
                break
        if not crossing and not inside_points:
            max_area = area
            max_coords = [ci, cj]
            max_square = square_sides
                    
            
print(max_area)
print(max_coords)

# Separate x and y for plotting
x, y = zip(*coords)
square_points = []
for a, b in max_square:
    square_points.append(a)
    square_points.append(b)

# Remove duplicates
square_points = list(dict.fromkeys(map(tuple, square_points)))

# Unpack x,y for plotting
mx, my = zip(*square_points)

# Plot all coordinates
plt.plot(x, y, color='blue', marker='o')
plt.plot(mx, my, color='red', marker='o')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Coordinates Plot')
plt.grid(True)
plt.show()
```
