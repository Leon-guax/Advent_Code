```python
from itertools import combinations
from tqdm import tqdm
```

## Part 1

```python
coords =[]
max_area = 0
with open('Day09_input.txt') as f:
     for line in f :
        coords.append([int(coord) for coord in line.rstrip('\n').split(',')])

for i, j in combinations(range(len(coords)), 2):
    ci = coords[i]
    cj = coords[j]
    x = abs(ci[0] -cj[0])+1
    y = abs(ci[1] -cj[1])+1
    area = x*y
    if area >max_area:
        max_area = area
print(max_area)
```
