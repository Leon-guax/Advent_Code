## Part 1

```python
def process_line_n(line, n_digits, verbose = False):
    verbose and print('\n--------\n', line)
    max_str=''
    
    begin_search_index = 0
    end_search_index = len(line) -n_digits
    
    for dig in range(n_digits-1 , -1, -1):
        end_search_index = len(line) -dig
        verbose and print('allowed substring = ', line[begin_search_index:end_search_index])        
        max_digit = max(line[begin_search_index:end_search_index])
        #first occurrence of first max digit within allowed search, rescaled
        max_digit_idx = line[begin_search_index:end_search_index].index(max_digit)+ begin_search_index
        max_str +=max_digit
        begin_search_index = max_digit_idx +1
        end_search_index = len(line) -dig
        
        verbose and print('max_str = ',max_str)        
    return int(max_str)
```

```python
verbose = False
jolt_sum = 0
with open('Day03_input.txt') as f:
     for line in f:
        jolt_sum +=process_line_n(line.rstrip('\n'),2 , verbose)

print(jolt_sum)
```
