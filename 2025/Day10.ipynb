```python
import numpy as np
import ast
from itertools import combinations
```

## Part 1

```python
n_steps = 0
with open('Day10_input.txt') as f:
     for line in f:
        lights, wires, jolt=line.rstrip('\n').replace(']', ']@').replace('{', '@{').split('@')
        lights= [i-1 for i, c in enumerate(lights) if c == '#']
        wires =  [ast.literal_eval(t) for t in wires.split()]
        wires = [(tup,) if isinstance(tup, int) else tup for tup in wires]
        
        set_list = [set(t) for t in wires]

        # Define the target pattern you want to reach
        found = False
        lights = set(lights)
        # Loop over combination sizes: 1, 2, 3
        for r in range(1, len(wires)+1):
            for combo in combinations(set_list, r):
                # XOR the sets (symmetric difference)
                result = set()
                for s in combo:
                    result ^= s  # symmetric difference
                if result == lights:
                    found = True
                    n_steps +=r
                    break
            if found:
                break
print(n_steps)
```

```python
def wires_to_deltas(wires, N):
    deltas = []
    for t in wires:
        row = [0]*N
        for x in t:      # t is a tuple of switch indices
            row[x] += 1  # you said "just add"
        deltas.append(row)

    return deltas
```
