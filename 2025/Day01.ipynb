## Part 2

```python
def part2(num, dir, pos, verbose = True):
    pw_2 =0
    if dir =='R':
        pos += num
        while pos >=100:
            pw_2+=1
            pos -=100
            verbose and print("    ", pos, '   pw_2 = ', pw_2)
    else:
        # if I don't cross or reach 0
        if pos > num:
            pos -= num
            return pos, pw_2
        # if I end up exactly on 0
        elif pos == num:
            pos -= num  # pos =0
            pw_2 +=1
            verbose and print("    ", pos, '   pw_2 = ', pw_2)
        # if I end below 0
        else:
            
            # if I start already on a 0, I am not counting an extra crossing
            if pos ==0:
                pw_2 -=1
                verbose and print("    ", pos, '   pw_2 = ', pw_2)
                
            pos -=num
            while pos <0:
                pw_2+=1
                pos+=100
                verbose and print("    ", pos, '   pw_2 = ', pw_2)
            # if I end up on -x00, this makes sure I also count the ending position
            if pos ==0:
                pw_2 +=1
                    
    return pos, pw_2
```

```python
pw_2 = 0

pos = 50   # current wheel 0..99
verbose = False

with open("Day01_input.txt") as f:
    for line in f:
        dir = line[0]
        num = int(line[1:].strip())
        verbose and print(line.replace('\n', ''), " pos = ", pos)
        pos, pw_2_part = part2(num, dir, pos, verbose)
        pw_2+=pw_2_part

    
print(f'part 2: {pw_2}')
```
