```python
import math
from itertools import combinations
import networkx as nx
```

## Part 1

```python
coords = []
with open('Day08_input.txt') as f:
     for line in f :
        coords.append([int(coord) for coord in line.rstrip('\n').split(',')])

pairs = []
for i, j in combinations(range(len(coords)), 2):
    d = math.dist(coords[i], coords[j])
    pairs.append((d, i, j))

pairs.sort()
shortest_edges = pairs[:1000]
edges = [ (i,j) for (d, i, j) in shortest_edges]

G = nx.Graph()
G.add_nodes_from(range(len(coords)))
G.add_edges_from(edges)
components = list(nx.connected_components(G))
components_sorted = sorted(components, key=len, reverse=True)

sizes = [len(c) for c in components_sorted]

print(sizes[0]*sizes[1]*sizes[2])
```

```python
# blatantly stolen from chatGPT, sorry not sorry

class DSU:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n
        self.components = n   # number of connected components
        
    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]
    
    def union(self, a, b):
        rootA = self.find(a)
        rootB = self.find(b)
        if rootA == rootB:
            return False  # already connected
        if self.rank[rootA] < self.rank[rootB]:
            self.parent[rootA] = rootB
        elif self.rank[rootA] > self.rank[rootB]:
            self.parent[rootB] = rootA
        else:
            self.parent[rootB] = rootA
            self.rank[rootA] += 1
        
        self.components -= 1
        return True
```

```python
coords = []
with open('Day08_input.txt') as f:
     for line in f :
        coords.append([int(coord) for coord in line.rstrip('\n').split(',')])

pairs = []
for i, j in combinations(range(len(coords)), 2):
    d = math.dist(coords[i], coords[j])
    pairs.append((d, i, j))
    
pairs.sort()
edges = [ (i,j) for (d, i, j) in pairs ]

dsu = DSU(len(coords))
last_edge = None
for i, j in edges:
    merged = dsu.union(i, j)  
    if merged:
        last_edge = (i, j)   
    if dsu.components == 1:
        break
    
i, j = last_edge
print(coords[i][0]*coords[j][0])
```
