## Part 1

```python
def update_char(line, pos, symbol='|'):
    return line[:pos] +symbol+line[pos+1:]
```

```python
Map = []
splitting = 0
prev_line = ''
with open('Day07_input.txt') as f:
     for line in f:
        if len(Map)==0:
            start = line.find('S')
        line = line.rstrip('\n')
        for idx,char in enumerate(line):
            if prev_line and prev_line[idx] in ['S', '|']:
                if char =='^': 
                    if line[idx-1]=='.':
                        line=update_char(line, idx-1)
                    if line[idx+1]=='.':
                        line=update_char(line, idx+1, symbol = '|')
                    splitting+=1
                    
                else:
                    line = update_char(line, idx)
        Map.append(line)
        prev_line = line
        

print(splitting)
```

## Part 2

```python
from collections import Counter

paths = Counter()
with open('Day07_input.txt') as f:
     for row, line in enumerate(f):
        if row==0:
            start = line.find('S')
            paths[start]=1
            continue
            
        new_paths = Counter()
        for path in paths:
            if line[path]=='^':
                new_paths[path-1]+=paths[path]
                new_paths[path+1]+=paths[path]
            else:
                new_paths[path]+=paths[path]
        paths=new_paths
        
print(sum(paths.values()))
```
