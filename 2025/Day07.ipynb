## Part 1

```python
def update_char(line, pos, symbol='S'):
    return line[:pos] +symbol+line[pos+1:]
```

```python
Map = []
splitting = 0
prev_line = ''
with open('Day07_input.txt') as f:
     for line in f:
        if len(Map)==0:
            start = line.find('S')
        line = line.rstrip('\n')
        for idx,char in enumerate(line):
            if prev_line and prev_line[idx] in ['S', '|']:
                if char =='^': 
                    if line[idx-1]=='.':
                        line=update_char(line, idx-1)
                    if line[idx+1]=='.':
                        line=update_char(line, idx+1, symbol = '|')
                    splitting+=1
                    
                else:
                    line = update_char(line, idx)
        Map.append(line)
        prev_line = line
        

print(splitting)
```

## Part 2

```python
Map = []
paths = []
prev_line = ''
with open('Day07_input.txt') as f:
     for row, line in enumerate(f):
        print(f'{row}) {len(paths)} ')
        if len(Map)==0:
            start = line.find('S')
            paths.append([start])
            line = line.rstrip('\n')
            Map.append(line)
            prev_line = line
            continue
            
        # I can always assume it's at least the second line
        new_paths = []
        for path in paths:
            if line[path[-1]]=='^':
                new_paths.append(path + [path[-1] -1])
                new_paths.append(path + [path[-1] +1])
            else:
                new_paths.append(path + [path[-1]])

        paths=new_paths
        
        
        line = line.rstrip('\n')
        Map.append(line)
        prev_line = line
        

print(len(paths))
```
