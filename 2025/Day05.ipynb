## Part 1

```python
ranges =[]
IDs = []
is_range = True
valid_sum = 0
with open('Day05_input.txt') as f:
    for line in f:
        if is_range and line !='\n':
            a, b= line.rstrip('\n').split('-')
            a = int(a)
            b = int(b)
            ranges.append([a,b])
        elif is_range and line =='\n':
            is_range = False          
        else: 
            ID=int(line.rstrip('\n'))
            IDs.append(ID)
            for rg in ranges:
                if ID>=rg[0] and ID <=rg[1]:
                    valid_sum +=1
                    break
            
print(valid_sum)
```

## Part 2

```python
def do_intersecate(l1, r1, l2, r2):
    return (l1<=l2 and l2<=r1) or (l1<=r2 and r2<=r1) 
```

```python
ranges =[]
with open('Day05_input.txt') as f:
    for line in f:
        new_range = True
        if line !='\n':
            a, b= line.rstrip('\n').split('-')
            a = int(a)
            b = int(b)            
            # check all the previous ranges. If there is intersection, just enlarge the existing one
            # given the while loop below, this initial operation should not be needed, but removing it changes the final value. Let's keep it as pre-processing step      
            for idx,rg in enumerate(ranges):
                if do_intersecate(a,b, rg[0], rg[1]):
                    ranges[idx][0] = min(a, rg[0])
                    ranges[idx][1] = max(b, rg[1])
                    new_range = False
                    break
            # if no existing one was enlarged, add a new range
            if new_range == True:
                ranges.append([a,b])    
        else:
            break

# merging all overlapping sets until it's no longer possible
old_rg_len = len(ranges)+1
while old_rg_len>len(ranges):
    old_rg_len= len(ranges)
    for idx,rg in enumerate(ranges):
        for jdx, sec_rg in enumerate(ranges[idx+1:]):
            if do_intersecate(rg[0], rg[1], sec_rg[0], sec_rg[1]):
                ranges[idx][0] = min(rg[0], sec_rg[0])
                ranges[idx][1] = max(rg[1], sec_rg[1])
                ranges[idx+jdx+1]= [0,-1] #everything merged gets updated with a sum-0 set
                break
    ranges =[x for x in ranges if x != [0, -1]] #sum-0 sets are removed. This makes the len(ranges) change, which means there were merged ranges
    

ranges_sum = 0
for rg in ranges:
    ranges_sum += (rg[1]-rg[0]+1)
    
print(ranges_sum)
```
