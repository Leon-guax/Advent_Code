## Part 1

```python
def safe_multiply(numbers):
    total = 1
    for num in numbers:
        total *= num
        # Check if total has overflowed (if it exceeds the max value for np.float64)
        if total > np.finfo(np.float64).max:
            raise OverflowError("Overflow occurred during multiplication")
    return total
```

```python
numbers =[]
with open('Day06_input.txt') as f:
     for line in f:
#         line = line.rstrip("\n")
        line_clean = re.findall(r'\d+', line)
        if line_clean !=[]:
            numbers.append([int(num) for num in line_clean])
        else:
            operations  = re.findall(r'[+*]', line)
            
numbers_T = np.array(numbers).T.tolist()

total_sum =0
for row in range(len(operations)):
    if operations[row]=='+':
        total_sum += sum(numbers_T[row])
    else:
        total_sum+= safe_multiply(numbers_T[row])
        
print(total_sum)

```

## Part 2

```python
def transpose_chars(scheme):
    new_scheme = []
    for idx in range(len(scheme[0])):
        new_line = []
        for jdx in range(len(scheme)):
            new_line.append(scheme[jdx][idx])
        new_scheme.append("".join(new_line))
    return new_scheme
```

```python
def separate_and_group_all(lst):
    result, group = [], []
    
    for item in lst:
        groups = item.split()
        if groups:  # If there are any non-empty groups
            group.extend(map(int, groups))
        elif group:  # If the current group is non-empty and we encounter an empty string
            result.append(group)
            group = []
    
    if group:  # Append the last group if it exists
        result.append(group)
    
    return result
```

```python
scheme =[]
with open('Day06_input.txt') as f:
     for line in f:
        line = line.rstrip("\n")
        scheme.append(line)


operations = scheme.pop(-1).split()
new_scheme = transpose_chars(scheme)
new_groups = separate_and_group_all(new_scheme)

total_sum =0
for row in range(len(operations)):
    if operations[row]=='+':
        total_sum += sum(new_groups[row])
    else:
        total_sum+= safe_multiply(new_groups[row])
        
print(total_sum)
```
