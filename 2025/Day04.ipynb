## Part 1

```python
def count_neigh(map, row, col, verbose = False):
    rolls = 0
    for x in [-1, 0, 1]:
        for y in [-1, 0, 1]:
            if not (x==0 and y ==0):
                verbose and print(f' for ball at ({row}, {col}), checking neighbour at row {row+y}, col {col+x}')
                if map[row +y][ col + x] =='@':
                   rolls +=1
    return rolls
```

```python
map = []
X_pos = []
size = 139

# add 1 layers of padding around the map
map.append('.'*(size+2))
with open('Day04_input.txt') as f:
     for row,line in enumerate(f):
        line = "."+line.rstrip("\n")+"."
        map.append(line)

map.append('.'*(size+2))

access_rolls = 0
for x in range(1,size+1):
    for y in range(1, size+1):
        if map[y][x]=='@' and count_neigh(map, y, x) <4:
            access_rolls+=1
            
print(access_rolls)
```
