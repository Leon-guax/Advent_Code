```python
# returns true if char is a digit char
def isdigit(char):
    return (char >= "0" and char <= "9")
```

```python
# given a line, and a tuple ['number_in_string', number]
# returns a line with all numbers in string form as numbers as int
def replace_str(line, str_to_digit):
    for digit in str_to_digit:
        line = line.replace(digit[0], str(digit[1]))
    return line
```

## Part 1

```python
sum =0


with open('Day1_input.txt') as f:
     for line in f:
         for char in line:
            if isdigit(char):
                sum = sum + 10*int(char)
                break
         for char in line[::-1]:
            if isdigit(char):
                sum = sum + int(char)
                break
         
print(sum)
```
