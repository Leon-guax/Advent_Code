```python
def isdigit(char):
    return (char >= "0" and char <= "9")
```

```python
def replace_str(line, str_to_digit):
    for digit in str_to_digit:
        line = line.replace(digit[0], str(digit[1]))
    return line
```

```python
sum =0


double_digit = [['oneight', 18],
                ['twone', 21],
                ['threeight', 38],
                ['fiveight', 58],
                ['eightwo', 82],
                ['eighthree', 83],
                ['nineight', 98]]
    


str_to_digit =[['one',1],
               ['two',2],
               ['three',3],
               ['four',4],
               ['five',5],
               ['six',6],
               ['seven',7],
               ['eight',8],
               ['nine',9]]

with open('Day1_input.txt') as f:
     for line in f:
         print(line)
         line = replace_str(line, double_digit)
         line = replace_str(line, str_to_digit)
         print(line +"\n")
         for char in line:
            if isdigit(char):
                sum = sum + 10*int(char)
                break
         for char in line[::-1]:
            if isdigit(char):
                sum = sum + int(char)
                break
         print(sum)

print(sum)
```
