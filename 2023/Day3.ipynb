```python
# given the whole map, and x_num_pos[], an int y_num_pos
# return true if there are only '.' symbols around those positions, and false otherwise
def is_isolated(map_pieces: list, x_num_pos:list , y_num_pos: int, x_size: int = 140, y_size: int = 140):
    flag = True
    
    if y_num_pos >0:
        flag = check_subset([map_pieces[y_num_pos -1][xdx] for xdx in x_num_pos])
        #print( 'y_num_pos >0 -' ,[map_pieces[y_num_pos -1][xdx] for xdx in x_num_pos] ,flag)
        if flag and x_num_pos[0] >0:
            #print('x_num_pos>0: ', [y_num_pos -1, y_num_pos+1],[x_num_pos[0]-1])
            flag = check_subset([map_pieces[ydx][x_num_pos[0]-1] for ydx in range(y_num_pos -1, y_num_pos+1)])
        if flag and x_num_pos[-1]< x_size -1:
            #print('x_num_pos[-1] <x_sixe: ',x_num_pos[-1], ' < ', x_size, '  ', [y_num_pos -1, y_num_pos+1],[x_num_pos[-1]+1])
            flag = check_subset([map_pieces[ydx][x_num_pos[-1]+1] for ydx in range(y_num_pos -1, y_num_pos+1)])
    
    if flag and y_num_pos <y_size-1:
        #print( 'y_num_pos y<y_size -' ,[map_pieces[y_num_pos +1][xdx] for xdx in x_num_pos] ,flag)
        #print([map_pieces[y_num_pos +1][xdx] for xdx in x_num_pos])
        flag = check_subset([map_pieces[y_num_pos +1][xdx] for xdx in x_num_pos])
        if flag and x_num_pos[0] >0:
            #print('[y_num_pos, y_num_pos+1 +1],[x_num_pos[0]-1] = ',[y_num_pos, y_num_pos+1 +1],[x_num_pos[0]-1])
            flag = check_subset([map_pieces[ydx][x_num_pos[0]-1] for ydx in range(y_num_pos, y_num_pos+1 +1)])
        if flag and x_num_pos[-1]< x_size -1:
            flag = check_subset([map_pieces[ydx][x_num_pos[-1]+1] for ydx in range(y_num_pos , y_num_pos+1 +1)])
    
    
    return flag
```

```python
# given an array of char
# returns false if any element is not a '.'
def check_subset(map_pieces_subset: list):
    for pos in map_pieces_subset:
            if pos !='.':
                return False
    return True
```

```python
def isdigit(char):
    return (char >= "0" and char <= "9")
```

```python
# given a line and an int x_position of first digit
# returns the array of positions and the value of that number
def find_numbers(map_pieces_line: list, x_position: int, x_size :int = 140) -> list, int:
    if (x_position == x_size-1) or (not isdigit(map_pieces_line[x_position +1])):
        return [x_position], int(map_pieces_line[x_position])
    
    #print(map_pieces_line, x_position)
    first_two_digits = int(map_pieces_line[x_position +1]) + 10*int(map_pieces_line[x_position])
    if x_position +2 <= x_size and isdigit(map_pieces_line[x_position +2]):
        return list(range(x_position, x_position+ 3)), int(map_pieces_line[x_position +2]) + 10*(first_two_digits )
    else:
        return list(range(x_position, x_position+ 2)), first_two_digits
  
```

```python
map_pieces = []
sum = 0

with open('Day3_input.txt') as f:
    for line in f:
        map_pieces.append(list(line.strip('\n')))
        
x_size = len(map_pieces[0])
y_size = len(map_pieces)

for ydx, line in enumerate(map_pieces):
    #print('-----------  Line ', ydx)
    for idx, char in enumerate(line):
        if isdigit(char) and (idx ==0 or not isdigit(line[idx-1]) ):
            x_idx_list, value = find_numbers(line, idx, x_size)
            #print('ydx, x_idx_list, value =  ', ydx, x_idx_list, value)
            if not is_isolated(map_pieces, x_idx_list, ydx, x_size, y_size):
                #print('adding' , value)
                sum = sum + value
                
print(sum)
```

```python
# given the whole map and an x_coord[] and y_coord, 
# returns an updated map with '.' in all those coordinates
def remove_numbers(map_pieces,  y_coord: int, x_coord: list):
    for x in list(x_coord):
        map_pieces[y_coord][x] = '.'
    return map_pieces
    
```

```python
# given a map line and an in x_coord of any digit
# returns the x_sition of the first digit of that block

def find_num_pos(map_line, x_coord):
    if x_coord == 0:
        return 0
    if x == 1:
        if isdigit(map_line[0]):
            return 0
        else:
            return 1
    if not isdigit (map_line[x_coord-1]):
        return x_coord
    if not isdigit(map_line[x_coord -2]):
        return x_coord -1
    else:
        return x_coord -2
```

```python
# given a whole map, and y_star_pos, x_star_pos of a '*' symbol
# if the '*' symbol has exactly two block numers neighbours, then
# it removes them from the map and returns the updated map and the two values

# seeing the input, I can always assume the * is an internal point
def exactly_two_neighb_num(map_pieces, y_star_pos, x_star_pos):
    block_numbers_to_add = []
    
    n_neighbours = 0
    temp_map = map_pieces
    
    x_square_coords = [-1, 0, 1, 1, 1, 0, -1, -1]
    y_square_coords = [-1, -1, -1, 0, 1, 1, 1, 0]
    
    for ydx, xdx in zip(y_square_coords, x_square_coords):
        if isdigit(temp_map[y_star_pos + ydx][x_star_pos + xdx]):
            first_x_pos = find_num_pos(temp_map[y_star_pos +ydx], x_star_pos + xdx)
            x_positions, temp = find_numbers(temp_map[y_star_pos +ydx], first_x_pos)
            block_numbers_to_add.append(temp)
            map_pieces = remove_numbers(temp_map, y_star_pos + ydx, x_positions)
            n_neighbours = n_neighbours +1

        
    
    if n_neighbours != 2:
        return map_pieces, 0, 0
    else:
        return temp_map, block_numbers_to_add[0], block_numbers_to_add[1]
```
