```python
def how_ranges_intersect(seed_range, rule):
    # rule 25 - 94 , range(80, 95) -> total
    
    if seed_range.stop < rule.start or seed_range.start > rule.stop+1:
        return False, 'none'
    
    if seed_range.start  >= rule.start and seed_range.stop > rule.stop+1:
        return True, 'bottom of range'
    
    if seed_range.start  < rule.start and seed_range.stop > rule.stop+1:
        return True, 'center of range'
    
    if seed_range.start  < rule.start and seed_range.stop <= rule.stop+1:
        return True, 'top of range'

    if seed_range.start  >= rule.start and seed_range.stop <= rule.stop+1:
        return True, 'total of range'
```

```python
def split_range(seed_range, rule, how):
    new_ranges = []
    match how:
        case 'bottom of range':
            new_ranges.append(range(seed_range.start + rule.value_diff, rule.stop + rule.value_diff+1 ))
            new_ranges.append(range(rule.stop+1, seed_range.stop))
            
        case 'center of range':
            new_ranges.append(range(rule.start + rule.value_diff, rule.stop +1 + rule.value_diff ) )
            new_ranges.append(range(seed_range.start, rule.start) )
            new_ranges.append(range(rule.stop +1, seed_range.stop) )  
            
        case 'top of range':
            new_ranges.append(range(rule.start + rule.value_diff, seed_range.stop + rule.value_diff))
            new_ranges.append(range(seed_range.start, rule.start))
            
        case 'total of range':
            new_ranges.append(range(seed_range.start + rule.value_diff, seed_range.stop + rule.value_diff))
        case _:
            print('how invalid')
    return new_ranges
```

```python
class NewRule:
    def __init__(self, digits):
        self.value_diff = digits[0] - digits[1]
        self.value_length = digits[2]
        self.start = digits[1]
        self.stop = digits[1]+digits[2]-1
```
